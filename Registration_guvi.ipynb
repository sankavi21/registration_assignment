```python
import re
import pandas as pd
import os
regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

def password_check(passwd):
    SpecialSym =['$', '@', '#', '%']
    val= True
    if len(passwd) < 5: 
        print('length should be at least 5') 
        val = False

    if len(passwd) > 16: 
        print('length should be not be greater than 16') 
        val = False

    if not any(char.isdigit() for char in passwd): 
        print('Password should have at least one numeral') 
        val = False

    if not any(char.isupper() for char in passwd): 
        print('Password should have at least one uppercase letter') 
        val = False

    if not any(char.islower() for char in passwd): 
        print('Password should have at least one lowercase letter') 
        val = False

    if not any(char in SpecialSym for char in passwd): 
        print('Password should have at least one of the symbols $@#') 
        val = False

    return val

def check_Email(email):
    if(re.fullmatch(regex, email)):
        return True
    else:
        print("Invalid Email")
        return False
def storeDetails(email,password):
    registerdetails = pd.DataFrame([[email,password]], columns=['email', 'password'])
    
    hdr = False  if os.path.isfile('register.csv') else True
    registerdetails.to_csv('register.csv', mode='a', header=hdr,index=False)
   
    return True

def resetPassword(email,password):
    resetdetails = pd.read_csv("register.csv")
    resetdetails.loc[resetdetails.email == email, 'password'] = password
    resetdetails.to_csv("register.csv", index=False)
    print("Password Reset Successfull")
    return True

def loginCheck(email,password):
    filename = 'register.csv'
    df = pd.read_csv(filename)
    for index, row in df.iterrows():
        if(row.email == email and row.password == password):
            return "success"
        elif(row.email == email and row.password != password):
            return "reset"
    return "failure"
    
while True:
    A=int(input("Register (or) login: press 1(or)2:      "))
    if A==1:
        B=input("enter your Email Id:")
        if(check_Email(B)):
            C=input("enter your password:")
            if(password_check(C)):
                flag=storeDetails(B,C)
                if(flag):
                    print("Successfully Registered")
            else:
                print("Registration Failed , Please Try Again")
                continue
    elif A==2:
        D=input("enter your Email Id:")
        E=input("enter your password:")
        login_flag=loginCheck(D,E)
        if(login_flag == "success"):
            print("SuccessFully Logged In")
        elif(login_flag == "reset"):
            print("Valid User , Password Incorrect")
            passwordreset=input("Do you want to reset password Y/N:")
            if(passwordreset == "Y" or passwordreset == "y"):
                newpassword=input("Enter New password:")
                resetPassword(D,newpassword)
            else:
                continue
                
        else:
            print("Not Valid User")
```
